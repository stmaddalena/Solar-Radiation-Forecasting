**Import libraries**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

**Load dataset**

```python
df = pd.read_csv('raw.csv')
df.head()
```

**Show information**

```python
df.info()
```

**Show describe**

```python
df.describe()
```

**Plot heat map**

```python
plt.figure(figsize=(30, 11))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

**Drop parameters**

```python
df.columns
```

```python
original = df.copy()
df.drop(columns=['date', 'tempmax', 'tempmin', 'feelslikemax',
       'feelslikemin', 'feelslike', 'dew', 'precipprob', 'preciptype', 'snow', 'snowdepth',
       'windspeed', 'winddir', 'sealevelpressure', 'cloudcover', 'visibility', 'severerisk', 'sunrise',
       'sunset', 'moonphase', 'description', 'icon', 'stations', 'PM25',
       'PM10', 'NO2', 'CO', 'weather'], axis=1, inplace=True)
```

**Check duplicate values**

```python
print('duplicated: %s'%df.duplicated().sum())
```

**Check outlier values**

```python
df.columns
```

```python
sns.set(style="whitegrid")
columns_to_plot = ['tempmax', 'tempmin', 'temp', 'humidity', 'precip', 'windgust', 'windspeed',
                   'sealevelpressure', 'cloudcover', 'solarradiation', 'PM10','NO2', 'SO2']
plt.figure(figsize=(25, 8))
sns.boxplot(data=df, orient="v", palette="Set2")
plt.title('Boxplots of outlier')
plt.ylabel('Value')
plt.xlabel('Parameter')
plt.xticks(rotation=90)
plt.show()
```

**Remove outlier values ​​and replace with NaN values**

```python
df.loc[(df['temp'] > 40.8) | (df['temp'] < 9.9), 'temp'] = np.nan
df.loc[(df['humidity'] > 100) | (df['humidity'] < 0), 'humidity'] = np.nan
df.loc[(df['precip'] > 248.6) | (df['precip'] < 0), 'precip'] = np.nan
df.loc[(df['precipcover'] > 100) | (df['precipcover'] < 0), 'precip'] = np.nan
df.loc[(df['windgust'] < 18), 'windgust'] = np.nan
df.loc[(df['O3'] > 121) | (df['O3'] < 0), 'O3'] = np.nan
df.loc[(df['SO2'] > 401) | (df['SO2'] < 0), 'SO2'] = np.nan
```

**Check missing values**

```python
df.isnull().sum(axis=0)
```

```python
plt.figure(figsize=(25, 8))
sns.barplot(data=df.isnull().sum().reset_index(), x='index', y=0)
plt.xticks(rotation=90)
plt.xlabel('Parameters')
plt.ylabel('Number of Missing Values')
plt.title('Missing Values per Column')
plt.show()
```

**Fill missing values (windgust)**

```python
df['windgust'].fillna(original.groupby('windspeed')['windgust'].transform('mean'), inplace=True)
df['windgust'].fillna(df['windgust'].mean(), inplace=True)
```

```python
fig = px.line(original, x=original.index, y='windgust', title='Comparison of <span style="color:red">windgust</span> between original data and filled data', labels={'x': 'Date', 'y': 'windgust'}, line_shape='linear')
fig.add_scatter(x=df.index, y=df['windgust'], mode='lines', name='filled data')
fig.add_scatter(x=original.index, y=original['windgust'], mode='lines', name='original data')
fig.show()
```

**Fill missing values (O3)**

```python
df['O3'].fillna(df.groupby('humidity')['O3'].transform('mean'), inplace=True)
df['O3'].fillna(df['O3'].mean(), inplace=True)
```

```python
fig = px.line(original, x=original.index, y='O3', title='Comparison of <span style="color:red">O3</span> between original data and filled data', labels={'x': 'Date', 'y': 'O3'}, line_shape='linear')
fig.add_scatter(x=df.index, y=df['O3'], mode='lines', name='filled data')
fig.add_scatter(x=original.index, y=original['O3'], mode='lines', name='original data')
fig.show()
```

**Fill missing values (SO2)**

```python
df['SO2'].fillna(df.groupby('O3')['SO2'].transform('mean'), inplace=True)
df['SO2'].fillna(df['SO2'].mean(), inplace=True)
```

```python
fig = px.line(original, x=original.index, y='SO2', title='Comparison of <span style="color:red">SO2</span> between original data and filled data', labels={'x': 'Date', 'y': 'SO2'}, line_shape='linear')
fig.add_scatter(x=df.index, y=df['SO2'], mode='lines', name='filled data')
fig.add_scatter(x=original.index, y=original['SO2'], mode='lines', name='original data')
fig.show()
```
