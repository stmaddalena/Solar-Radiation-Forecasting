**Import libraries**

```python
import pandas as pd
import missingno as msno
import seaborn as sns
import matplotlib.pyplot as plt
```

**Load dataset 1**

```python
bangkok_1 = pd.read_csv('bangkok 1.csv')
nakhonpathom_1 = pd.read_csv('nakhonpathom 1.csv')
nonthaburi_1 = pd.read_csv('nonthaburi 1.csv')
pathumthani_1 = pd.read_csv('pathumthani 1.csv')
samutprakan_1 = pd.read_csv('samutprakan 1.csv')
samutsakhon_1 = pd.read_csv('samutsakhon 1.csv')
```

**Load dataset 2**

```python
bangkok_2 = pd.read_csv('bangkok 2.csv')
nakhonpathom_2 = pd.read_csv('nakhonpathom 2.csv')
nonthaburi_2 = pd.read_csv('nonthaburi 2.csv')
pathumthani_2 = pd.read_csv('pathumthani 2.csv')
samutprakan_2 = pd.read_csv('samutprakan 2.csv')
samutsakhon_2 = pd.read_csv('samutsakhon 2.csv')
```

**Change the format of a datetime**

```python
bangkok_2['date'] = pd.to_datetime(bangkok_2['date'])
bangkok_2['date'] = bangkok_2['date'].dt.strftime('%Y-%m-%d')

nakhonpathom_2['date'] = pd.to_datetime(nakhonpathom_2['date'])
nakhonpathom_2['date'] = nakhonpathom_2['date'].dt.strftime('%Y-%m-%d')

nonthaburi_2['date'] = pd.to_datetime(nonthaburi_2['date'])
nonthaburi_2['date'] = nonthaburi_2['date'].dt.strftime('%Y-%m-%d')

pathumthani_2['date'] = pd.to_datetime(pathumthani_2['date'])
pathumthani_2['date'] = pathumthani_2['date'].dt.strftime('%Y-%m-%d')

samutprakan_2['date'] = pd.to_datetime(samutprakan_2['date'])
samutprakan_2['date'] = samutprakan_2['date'].dt.strftime('%Y-%m-%d')

samutsakhon_2['date'] = pd.to_datetime(samutsakhon_2['date'])
samutsakhon_2['date'] = samutsakhon_2['date'].dt.strftime('%Y-%m-%d')
```

**Rename column**

```python
bangkok_1.rename(columns={'datetime': 'date', 'conditions': 'weather'}, inplace=True)
nakhonpathom_1.rename(columns={'datetime': 'date', 'conditions': 'weather'}, inplace=True)
nonthaburi_1.rename(columns={'datetime': 'date', 'conditions': 'weather'}, inplace=True)
pathumthani_1.rename(columns={'datetime': 'date', 'conditions': 'weather'}, inplace=True)
samutprakan_1.rename(columns={'datetime': 'date', 'conditions': 'weather'}, inplace=True)
samutsakhon_1.rename(columns={'datetime': 'date', 'conditions': 'weather'}, inplace=True)

bangkok_2.rename(columns={' pm25': 'PM25', ' pm10': 'PM10', ' o3': 'O3', ' no2': 'NO2', ' so2': 'SO2', ' co': 'CO'}, inplace=True)
nakhonpathom_2.rename(columns={' pm25': 'PM25', ' pm10': 'PM10', ' o3': 'O3', ' no2': 'NO2', ' so2': 'SO2', ' co': 'CO'}, inplace=True)
nonthaburi_2.rename(columns={' pm25': 'PM25', ' pm10': 'PM10', ' o3': 'O3', ' no2': 'NO2', ' so2': 'SO2', ' co': 'CO'}, inplace=True)
pathumthani_2.rename(columns={' pm25': 'PM25', ' pm10': 'PM10', ' o3': 'O3', ' no2': 'NO2', ' so2': 'SO2', ' co': 'CO'}, inplace=True)
samutprakan_2.rename(columns={' pm25': 'PM25', ' pm10': 'PM10', ' o3': 'O3', ' no2': 'NO2', ' so2': 'SO2', ' co': 'CO'}, inplace=True)
samutsakhon_2.rename(columns={' pm25': 'PM25', ' pm10': 'PM10', ' o3': 'O3', ' no2': 'NO2', ' so2': 'SO2', ' co': 'CO'}, inplace=True)
```

```python
bangkok_2['NO2'] = pd.to_numeric(bangkok_2['CO'], errors='coerce')
bangkok_2['SO2'] = pd.to_numeric(bangkok_2['SO2'], errors='coerce')
bangkok_2['PM25'] = pd.to_numeric(bangkok_2['PM25'], errors='coerce')
bangkok_2['PM10'] = pd.to_numeric(bangkok_2['PM10'], errors='coerce')
bangkok_2['O3'] = pd.to_numeric(bangkok_2['O3'], errors='coerce')
bangkok_2['CO'] = pd.to_numeric(bangkok_2['CO'], errors='coerce')

nakhonpathom_2['NO2'] = pd.to_numeric(nakhonpathom_2['CO'], errors='coerce')
nakhonpathom_2['SO2'] = pd.to_numeric(nakhonpathom_2['SO2'], errors='coerce')
nakhonpathom_2['PM25'] = pd.to_numeric(nakhonpathom_2['PM25'], errors='coerce')
nakhonpathom_2['PM10'] = pd.to_numeric(nakhonpathom_2['PM10'], errors='coerce')
nakhonpathom_2['O3'] = pd.to_numeric(nakhonpathom_2['O3'], errors='coerce')
nakhonpathom_2['CO'] = pd.to_numeric(nakhonpathom_2['CO'], errors='coerce')

nonthaburi_2['NO2'] = pd.to_numeric(nonthaburi_2['CO'], errors='coerce')
nonthaburi_2['SO2'] = pd.to_numeric(nonthaburi_2['SO2'], errors='coerce')
nonthaburi_2['PM25'] = pd.to_numeric(nonthaburi_2['PM25'], errors='coerce')
nonthaburi_2['PM10'] = pd.to_numeric(nonthaburi_2['PM10'], errors='coerce')
nonthaburi_2['O3'] = pd.to_numeric(nonthaburi_2['O3'], errors='coerce')
nonthaburi_2['CO'] = pd.to_numeric(nonthaburi_2['CO'], errors='coerce')

pathumthani_2['NO2'] = pd.to_numeric(pathumthani_2['CO'], errors='coerce')
pathumthani_2['SO2'] = pd.to_numeric(pathumthani_2['SO2'], errors='coerce')
pathumthani_2['PM25'] = pd.to_numeric(pathumthani_2['PM25'], errors='coerce')
pathumthani_2['PM10'] = pd.to_numeric(pathumthani_2['PM10'], errors='coerce')
pathumthani_2['O3'] = pd.to_numeric(pathumthani_2['O3'], errors='coerce')
pathumthani_2['CO'] = pd.to_numeric(pathumthani_2['CO'], errors='coerce')

samutprakan_2['NO2'] = pd.to_numeric(samutprakan_2['CO'], errors='coerce')
samutprakan_2['SO2'] = pd.to_numeric(samutprakan_2['SO2'], errors='coerce')
samutprakan_2['PM25'] = pd.to_numeric(samutprakan_2['PM25'], errors='coerce')
samutprakan_2['PM10'] = pd.to_numeric(samutprakan_2['PM10'], errors='coerce')
samutprakan_2['O3'] = pd.to_numeric(samutprakan_2['O3'], errors='coerce')
samutprakan_2['CO'] = pd.to_numeric(samutprakan_2['CO'], errors='coerce')

samutsakhon_2['NO2'] = pd.to_numeric(samutsakhon_2['CO'], errors='coerce')
samutsakhon_2['SO2'] = pd.to_numeric(samutsakhon_2['SO2'], errors='coerce')
samutsakhon_2['PM25'] = pd.to_numeric(samutsakhon_2['PM25'], errors='coerce')
samutsakhon_2['PM10'] = pd.to_numeric(samutsakhon_2['PM10'], errors='coerce')
samutsakhon_2['O3'] = pd.to_numeric(samutsakhon_2['O3'], errors='coerce')
samutsakhon_2['CO'] = pd.to_numeric(samutsakhon_2['CO'], errors='coerce')
```

**Merged data**

```python
bangkok_1 = bangkok_1.merge(bangkok_2, left_on='date', right_on='date', how='left')
nakhonpathom_1 = nakhonpathom_1.merge(nakhonpathom_2, left_on='date', right_on='date', how='left')
nonthaburi_1 = nonthaburi_1.merge(nonthaburi_2, left_on='date', right_on='date', how='left')
pathumthani_1 = pathumthani_1.merge(pathumthani_2, left_on='date', right_on='date', how='left')
samutprakan_1 = samutprakan_1.merge(samutprakan_2, left_on='date', right_on='date', how='left')
samutsakhon_1 = samutsakhon_1.merge(samutsakhon_2, left_on='date', right_on='date', how='left')
```
